```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_file(name = None): 
    if name is not None: 
        score = []
        with open(name) as f: 
            line = f.readline() 
            s = float(line)
            score.append(s)
            while line: 
                line = f.readline()
                if line =='':
                    break
                s = float(line)
                score.append(s)
        return score
```

```python
score_1 = read_file('data_model_1.txt')
```

```python
score_1_dec_noise= read_file('data_model_1_dec.txt')
```

```python
score_2 = read_file('data_model_2.txt')
```

```python
score_3 = read_file('data_model_3.txt')
```

```python
score_4 = read_file('data_model_4.txt')
```

```python
x = [i+1 for i in range(2000)]  
```

```python
running_avg_1= np.zeros(len(score_1))
```

```python
running_avg_2 = np.zeros(len(score_1))
running_avg_3 = np.zeros(len(score_1))
running_avg_4 = np.zeros(len(score_1))
running_avg_5 = np.zeros(len(score_1))
```

```python
for i in range(len(running_avg_1)):
    running_avg_1[i] = np.mean(score_1[max(0, i-50):(i+1)])
    running_avg_2[i] = np.mean(score_2[max(0, i-50):(i+1)])
    running_avg_3[i] = np.mean(score_3[max(0, i-50):(i+1)])
    running_avg_4[i] = np.mean(score_4[max(0, i-50):(i+1)])
    running_avg_5[i] = np.mean(score_1_dec_noise[max(0, i-50):(i+1)])   
```

```python
plt.plot(x, running_avg_1,label='critic and actor lr = 0.001')
plt.plot(x, running_avg_2,label='critic and actor lr = 0.005')
plt.plot(x, running_avg_3,label='critic and actor lr = 0.0025')
plt.plot(x, running_avg_4,label='critic and actor lr = 0.00075')
plt.xlabel('Episode')
plt.ylabel('Scores Average')
plt.title('Running average of previous 50 scores')
plt.legend()
```

```python
plt.plot(x, running_avg_1,label='model with constant noise')
plt.plot(x, running_avg_5,label='model with decreasing noise')
plt.xlabel('Episode')
plt.ylabel('Scores Average')
plt.title('Running average of previous 50 scores')
plt.legend()
```
