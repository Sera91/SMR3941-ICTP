```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_file(name = None): 
    if name is not None: 
        score = []
        with open(name) as f: 
            line = f.readline() 
            s = float(line)
            score.append(s)
            while line: 
                line = f.readline()
                if line =='':
                    break
                s = float(line)
                score.append(s)
        return score
```

```python
score_1 = read_file(name="data/doc1.txt")
```

```python
score_2 = read_file(name="data/doc3.txt")
```

```python
x = [i+1 for i in range(2000)]  
```

```python
running_avg_1= np.zeros(len(score_1))
running_avg_2 = np.zeros(len(score_1))
```

```python
for i in range(len(running_avg_1)):
    running_avg_1[i] = np.mean(score_1[max(0, i-50):(i+1)])
    running_avg_2[i] = np.mean(score_2[max(0, i-50):(i+1)])
    
```

```python
plt.plot(x, running_avg_1,label='Model with OU-noise')
plt.plot(x, running_avg_2,label='Model with decreasing normal noise')
plt.xlabel('Episode')
plt.ylabel('Scores Average')
plt.title('Running average of previous 50 scores')
plt.legend()
```
